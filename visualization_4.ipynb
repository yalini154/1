```python
import matplotlib.pyplot as plt
blood_suger_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_suger_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

type = [blood_suger_men,blood_suger_women]
colors = ['g', 'r']
label = ['men', 'woman']
bins = [80, 100, 125, 150]
plt.xlabel = ("Blood Suger Range")
plt.ylabel("Total no. of patients")
plt.hist(type, bins=bins, rwidth=0.95, color=colors, label=label, orientation="horizontal")

plt.title("Blood Suger Level Chart")
plt.legend()
plt.show()
```
