```python
def printnumber(n):
  literation = 0
  print("the total number entered my user is", n)
  literation+=1
  print("total literation done by computer is",literation,"\n")
printnumber(10)
printnumber(200)
```

```python
def OnTime(n):
  iteration=0
  for i in range(1, n+1):
    iteration+=1
  print("Whan n is",n," Iteratons = ",iteration)
OnTime(10)
OnTime(200)
```

```python
def test(n):
  iteration = 0
  for i in range(0,n):
    for j in range(0,n):
      print("*",end="")
      iteration+=1
    print("")
  print("\nwhen n is",n,"iteration =",iteration)
test(10)
test(200)
```
