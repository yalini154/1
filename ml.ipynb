```python
import nltk
from nltk.chat.util import Chat, reflections

reflections = {
    "i am" : "you are",
    "i was" : "you were",
    "i"  :  "you",
    "i'm"  :  "you are",
    "i'd"  :  "you would",
    "i've"  :  "you have",
    "i'll"  :  "you will",
    "my"  :  "your",
    "you are"  :  "I am",
    "you were"  :  "I was",
    "you've"  :  "I have",
    "you'll"  :  "I will",
}

pairs = [
    [
    r"my name is (.*)",
    ["Hello %1, How are you today"]
],
[
    r"hi|hey|hello",
    ["Hello", "Hey there"]
],
[
    r"how are you ?",
    ["I'm doing goodnHow are you"]
]

]
def chat():
  print("Hi! I am a chatbot created by yalini. for your service")
  chat = Chat(pairs, reflections)
  chat.converse()

if __name__ == "__main__":
  chat()
```
